```python
class student:
    clg_name="RGMCET"
    address="Nndyal"
    clg_code="RGIT"
    no_students=0
    student_details={}
    def __init__(self,phone,address,marks1,marks2,marks3):
        self.name=self.name_validate()
        self.email=self.email_validate()
        self.phone=phone
        self.address=address
        self.marks1=marks1
        self.marks2=marks2
        self.marks3=marks3
        self.id_generate(self)
        self.calculateTotal(marks1,marks2,marks3)
        self.concate_student_details(self)


    def name_validate(self):
        name=input("Enter student name: ")
        if not(3<=len(name)<=16):
            print("Student name should be min 3 and max 16 characters")
            self.name_validate()
        elif not(name.isalpha()):
            print("Student name should be in alphabet characters")
            self.name_validate()
        else:
            return name.upper()
        
    def email_validate(self):
        email=input("Enter email id:")
        if not(8<=len(email)<=25):
            print("Email should contain min 8 and max 25 characters")
            self.email_validate()
        elif not(email[0].isalpha()):
            print("Email first character should be alphabet")
            self.email_validate()
        elif not(email.count('@')==1 and  '.'  in email[email.index('@'):] and 1<=email.count('.')<=3):
            print("Fghgjghjgh")
            self.email_validate()
        elif len([i for i in email if not('A'<=i<='Z' or 'a'<=i<='z' or '0'<=i<='9' or i in '@.') ]):
            print("EMNJKNJ")
            self.email_validate()
        else:
            return email

    @classmethod
    def id_generate(cls,self):
        cls.no_students+=1
        self.id=123434+cls.no_students

    @classmethod
    def concate_student_details(cls,self):
        if self.id in cls.student_details:
            cls.student_details[self.id]+=self
        else:
            cls.student_details[self.id]=self
    
    @classmethod
    def display_stuDetails(cls):
        print("Students Details".center(50,'*'))
        for i in cls.student_details:
            print("Roll Number",cls.student_details[i].id)
            print("Student Name",cls.student_details[i].name)
            print("Total Marks",cls.student_details[i].Total_marks)
            print("Percentage",cls.student_details[i].percentage)
            print('\t')
    

    @classmethod
    def display_specific_stuDetails(cls):
        stuID=int(input("enter student id whoose details have to display:"))
        if stuID in cls.student_details:
            print(cls.student_details[stuID].__dict__)
        else:
            cls.invalid_id()
        

    def calculateTotal(self,marks1,marks2,marks3):
        self.percentage=((marks1+marks2+marks3)/(3*100))*100
        self.Total_marks=marks1+marks2+marks3

    def modify_marks(self):
        n=int(input("choose an option:\n1.marks1\n2.marks2\n3.marks3\n4.exit"))
        match n:
            case 1:
                self.marks1=int(input("enter marks:"))
            case 2:
                self.marks2=int(input("enter marks:"))
            case 3:
                self.marks3=int(input("enter marks:"))
            case _:
                return
        return self.modify_marks()
    
    def change_phone_number(self):
        self.phone=int(input("enter new mobile number:"))

    @staticmethod
    def invalid_id():
        print("Invalid student ID")







student1=student(9014983958,"Kadapa",90,85,80)
# student2=student("niranjangujjula@gmail.com",9014983959,"Kadapa",92,80,87)
# student3=student("lalithasana@gmail.com",9014983968,"Kadapa",93,82,86)

# print(student1.__dict__)
# print(student2.__dict__)
# print(student3.__dict__)




def main(cls):
    stu_id=int(input("enter student id:"))
    if stu_id in cls.student_details:
        n=int(input("choos an option:\n1.display all students details\n2.display specific student details\n3.modify marks\n4.change mobile\n5.exit"))
        self=cls.student_details[stu_id]
        match n:
            case 1:
                cls.display_stuDetails()
            case 2:
                cls.display_specific_stuDetails()
            case 3:
                self.modify_marks()
            case 4:
                self.change_phone_number()
            case _:
                return
    else:
        cls.invalid_id()
        main(cls)
    
    

main(student)

print(student1.__dict__)
# print(student2.__dict__)
# print(student3.__dict__)

print(student.student_details)

    







```

```python
class company:
    c_name="DELOITTE"
    c_location="HYDERBAD"
    c_code="DEL1234"
    employee_details={}
    no_employees=0

    def __init__(self,e_name,e_phone,e_sal,e_domain):
        self.e_name=e_name
        self.e_phone=e_phone
        self.e_sal=e_sal
        self.e_domain=e_domain
        self.emp_id_creation()
        self.adding_employee_details(self)

    def emp_id_creation(self):
        self.no_employees+=1
        self.e_id=1000+self.no_employees

    @classmethod
    def adding_employee_details(cls,self):
        cls.employee_details[self.e_id]=self

    def display_employee_ditails(self):
        id=int(input("Enter the employee id: "))
        if id in self.employee_details:
            print(self.employee_details[id].__dict__)
        else:
            self.invalid_id()

    @classmethod
    def hike_Salary(cls,self):
        id=int(input("Enter the employee id: "))
        if id in cls.employee_details:
            cls.employee_details[id].e_sal=(self.e_sal+(self.e_sal)*(4//100))*12
            print("Hike added successfully")
        else:
            self.invalid_id()


    @staticmethod
    def invalid_id():
        print("Employee is not present")

narendra=company('narendra',9014983958,25000,'developer')

def main(cls):
    id=int(input("enter student id:"))
    if id in cls.employee_details:
        n=int(input("choos an option:\n1.display employee details\n2.update hike of salary\n3.exit"))
        self=cls.employee_details[id]
        match n:
            case 1:
                cls.display_employee_ditails(self)
            case 2:
                cls.hike_Salary(self)
            case _:
                return
    else:
        cls.invalid_id()

main(company)

print(company.employee_details[1001].__dict__)


    
            
    
        

```
