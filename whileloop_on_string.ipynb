```python
#1.wap to check the email is valid or not if valid display valid or else display invalid.
#constraints are 1.no special characters will allow except '@' and '.' 2.exactly one @ will be there
#3.after @ minimum one '.' and maximum two '.' will allow.
s='narendra@gmail.com'
a=0
count1=0
while a<len(s):
    if not('A'<=s[a]<='Z' or 'a'<=s[a]<='z' or '0'<=s[a]<='9' or s[a] in '@.'):
        count1+=1
    a+=1
if count1==0:
    if '@' in s and '.' in s:
        b=0
        count2=0
        pos=None
        while b<len(s):
            if s[b] == '@':
                pos=b
                count2+=1
            b+=1
        #print(pos)
        if count2==1:
            if '.' in s[pos::]:
                count3=0
                while pos<len(s):
                    if s[pos] in '.':
                        count3+=1
                    pos+=1
                if 1<=count3<=2:
                    print('valid')
                else:
                    print('invalid')
            else:
                print('in valid')
        else:
            print('invalid')
    else:
        print('invalid')
else:
    print('invalid')
```
